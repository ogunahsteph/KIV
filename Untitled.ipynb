### Load the dataset to explore

### Modify Datess

```python
def encode_dates(df, feature):
    df = df.copy()
    
    df[feature] = pd.to_datetime(df[feature])
    
    df[feature + "_year"] = df[feature].apply(lambda x: x.year)
    df[feature + "_month"] = df[feature].apply(lambda x: x.month)
    df[feature + "_day"] = df[feature].apply(lambda x: x.day)
    
    # df[+ "_hour"] = df[feature].apply(lambda x: x.hour)
    
    
    df = df.drop(feature, axis=1)
    
    return df
```

### Data Preprocessing

```python
def pre_processor(df):
     df = df.copy()
    
    # Drop id column
     df = df.drop('id', axis=1)
    
    # Drop use and tags columns (avoiding NLP)
     df = df.drop(['use', 'tags'], axis=1)
    
    # Drop country and date columns (redundant information)
     df = df.drop(['country_code','currency','partner_id','date'], axis=1)
    
    # Drop region column (high-cardinality)
     df = df.drop('region', axis=1)

     df = encode_dates(df, feature='posted_time')
     df = encode_dates(df, feature='disbursed_time')
     df = encode_dates(df, feature='funded_time')
    #Create Funding gap feature
     df['funding_gap']=(df['loan_amount']-df['funded_amount'])/df['loan_amount']*100
     

     return df
```

```python
df=pre_processor(df)
df
```

```python
# Get unique countries in the DataFrame
unique_countries = df["country"].unique()

# Calculate the number of rows and columns for subplots
num_countries = len(unique_countries)
num_cols = 2  # Number of columns for subplots
num_rows = (num_countries + num_cols - 1) // num_cols

# Create subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(18, 7*num_rows))

# Loop over each country and plot in subplots
for i, country in enumerate(unique_countries):
    row = i // num_cols
    col = i % num_cols
    
    # Filter the DataFrame for the current country
    country_df = df[df["country"] == country]
    
    # Plot the KDE
    sns.countplot(data=country_df, x="repayment_interval", hue="sector", fill=True, ax=axes[row, col])
    
    # Set the title and labels
    axes[row, col].set_title(f"Distribution of loan amount by repayment_interval in {country}", size=15)
    axes[row, col].set_xlabel("Repayment Interval", size=15)
    axes[row, col].set_ylabel("Loan Amount", size=15)
    axes[row, col].tick_params(axis='x', labelsize=12)
    axes[row, col].tick_params(axis='y', labelsize=12)
    
    # Add legend
    axes[row, col].legend(title="Sector", fontsize=12, title_fontsize=12)
    
# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Calculate the mean funding gap for each sector
mean_funding_gap_by_sector = df.groupby("sector")["funding_gap"].mean().reset_index()

# Plot a bar chart
plt.figure(figsize=(7, 4))
plt.gca().set_facecolor('lightgrey')
sns.barplot(data=mean_funding_gap_by_sector, y="sector", x="funding_gap", palette="viridis")

# Set the title and labels
plt.title("Mean Funding Gap by Sector", size=10)
plt.xlabel("Mean Funding Gap (%)", size=10)
plt.ylabel("Sector", size=15)
plt.xticks(rotation=0, ha="right", size=12)
plt.yticks(size=10)

# Show the plot
plt.show()

```
